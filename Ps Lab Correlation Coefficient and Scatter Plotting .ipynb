```python
import matplotlib.pyplot as plt
import numpy as np
height =np.random.normal(170,10,100)
weight=np.random.normal(70,15,100)
plt.scatter(height,weight)
plt.xlabel("Height(cm)")
plt.ylabel("Weight(kg)")
plt.title("Height vs Weight Scatter Plot")
plt.show()
```

```python
import numpy as np
height =np.random.normal(170,10,100)
weight=np.random.normal(70,15,100)
correlation=np.corrcoef(height,weight)[0,1]
print("Pearson Correlation Coefficient:",correlation)
```

```python
import numpy as np
height =np.random.normal(170,10,100)
weight=np.random.normal(70,15,100)
age=np.random.normal(30,5,100)
correlation_matrix=np.corrcoef([height,weight,age])
print("Correlation Matrix:\n",correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import numpy as np
height =np.random.normal(170,10,100)
weight=np.random.normal(70,15,100)
age=np.random.normal(30,5,100)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(height,weight)
plt.xlabel("Height(cm)")
plt.ylabel("Weight(kg)")
plt.title("Height vs Weight Scatter Plot")

plt.subplot(1,3,2)
plt.scatter(height,age)
plt.xlabel("Height(cm)")
plt.ylabel("Age")
plt.title("Height vs Age Scatter Plot")

plt.subplot(1,3,3)
plt.scatter(weight,age)
plt.xlabel("Weight(kg))")
plt.ylabel("Age")
plt.title("weight vs Age Scatter Plot")

plt.tight_layout()
plt.show()
```
