```python
import matplotlib.pyplot as plt
import numpy as np
study_hours =np.array([2,4,6,8,1,5,7,3,9,10])
exam_score=np.array([50,60,70,85,40,65,78,55,90,95])
sleep_hours=np.array([8,7,6,5,9,6.5,5.5,7.5,4,3])
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(study_hours,exam_score)
plt.xlabel("study_hours")
plt.ylabel("exam_score")
plt.title("study_hours vs exam_score Scatter Plot")

plt.subplot(1,3,2)
plt.scatter(study_hours,sleep_hours)
plt.xlabel("study_hours")
plt.ylabel("sleep_hours")
plt.title("study_hours vs sleep_hours Scatter Plot")

plt.subplot(1,3,3)
plt.scatter(exam_score,sleep_hours)
plt.xlabel("exam_score)")
plt.ylabel("sleep_hours")
plt.title("exam_score vs sleep_hours Scatter Plot")

plt.tight_layout()
plt.show()
```

```python
correlation=np.corrcoef(study_hours,exam_score)[0,1]
print("Pearson Correlation Coefficient:",correlation)
```

```python
correlation=np.corrcoef(study_hours,sleep_hours)[0,1]
print("Pearson Correlation Coefficient:",correlation)
```
